```python
## Which department has seen the most attrition?
```

```python
data['Department'].unique()
```

```python
data.groupby('Department')['Attrition'].value_counts()
```

```python
data.query('Attrition=="Yes"').groupby('Department')['Attrition'].value_counts()
```

```python
data['Attrition_mapped'] = (data['Attrition'] == "Yes").map({True:1, False:0})
```

```python
data.head()
```

```python
P = data.query('Attrition_mapped==1').groupby('Department').agg({'Attrition_mapped':np.sum})
P['Percentage'] = 100 * P['Attrition_mapped']  / P['Attrition_mapped'].sum()
P
```

Research & Development department has the highest attrition rate

```python
## Which Job Role has seen the most attrition?
```

```python
data['JobRole'].unique()
```

```python
P = data.query('Attrition_mapped==1').groupby('JobRole').agg({'Attrition_mapped':np.sum})
P['Percentage'] = 100 * P['Attrition_mapped']  / P['Attrition_mapped'].sum()
P.sort_values('Percentage', ascending=False)
```

Lab Technician job role was the most impacted by attrition and has the highest attrition rate

```python
## Does income parity across departments have an effect on attrition?
```

```python
data.groupby('Department').agg({'MonthlyIncome':np.mean})
```

```python
P = data.query('Attrition_mapped==1').groupby('Department').agg({'Attrition_mapped':np.sum, 'MonthlyIncome':np.mean})
P['Percentage'] = 100 * P['Attrition_mapped']  / P['Attrition_mapped'].sum()
P
```

```python
P = data.query('Attrition_mapped==0').groupby('Department').agg({'Attrition_mapped':'count', 'MonthlyIncome':np.mean})
P['Percentage'] = 100 * P['Attrition_mapped']  / P['Attrition_mapped'].sum()
P
```

Employees within the same department who attrite have lesser monthly income compared to those who don't
