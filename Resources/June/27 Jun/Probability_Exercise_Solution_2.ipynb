```python
# 1)A company has a team of 10 employees, consisting of 6 women and 4 men.
# The HR department needs to randomly select 3 employees to participate in a leadership training program. What is the probability that all 3 employees selected are women?

total_employees = 10
women = 6
probability = (women / total_employees) * ((women - 1) / (total_employees - 1)) * ((women - 2) / (total_employees - 2))
print("Probability of selecting all 3 women:", round(probability,3))
```

```python
# 2)A retail store has a stock of 50 shirts, consisting of 30 small size and 20 medium size shirts.
# The store owner needs to randomly select 5 shirts to display as samples. What is the probability that exactly 3 of the selected shirts are small size?

total_shirts = 50
small_size_shirts = 30
probability = (small_size_shirts / total_shirts) * ((small_size_shirts - 1) / (total_shirts - 1)) * ((small_size_shirts - 2) / (total_shirts - 2)) * ((total_shirts - small_size_shirts) / (total_shirts - 3)) * ((total_shirts - small_size_shirts - 1) / (total_shirts - 4))
print("Probability of selecting exactly 3 small size shirts:", round(probability,3))
```

```python
# 4)A company has to select an AI/ML Engineer from a list of 50 people of which 20 are women and 30 are men, and 10 are knowledgeable in API.
# Fifteen of them have more than 2 years of industrial experience and the remaining 35 do not.
# What is the likelihood that the employer will choose a woman with more than two years of industrial experience who is familiar with API?

P_Woman = 20 / 50                  # Probability of selecting a woman
P_Experience_gt_2_years = 15 / 50  # Probability of having more than two years of industrial experience
P_API = 10 / 50                    # Probability of being knowledgeable in API

# Calculate the likelihood of choosing a woman with more than two years of industrial experience who is familiar with API
likelihood = P_Woman * P_Experience_gt_2_years * P_API

print("Likelihood of choosing a woman with more than two years of industrial experience who is familiar with API:", likelihood)
```

```python
# 5)In a factory a product is manufactured by 3 machines A, B, and C. They produce 25%, 35% and 40% of the total products.
# A product is selected at random. Then find the probability that it is defective if the given machines A, B, and C produce 5%, 4%
# and 2% defective items.

P_A = 0.25                  # Probability of machine A producing the product
P_B = 0.35                  # Probability of machine B producing the product
P_C = 0.40                  # Probability of machine C producing the product
P_Defective_given_A = 0.05  # Probability of a defective item produced by machine A
P_Defective_given_B = 0.04  # Probability of a defective item produced by machine B
P_Defective_given_C = 0.02  # Probability of a defective item produced by machine C

# Calculate the probability that a randomly selected product is defective
P_Defective = P_Defective_given_A * P_A + P_Defective_given_B * P_B + P_Defective_given_C * P_C

print("Probability that a randomly selected product is defective:", P_Defective)
```
