```python
#Question 2
# Calculate the mean salary
mean_salary = df['Annual Salary($)'].mean()

# Calculate the median salary
median_salary = df['Annual Salary($)'].median()

# Calculate the range of salaries
salary_range = df['Annual Salary($)'].max() - df['Annual Salary($)'].min()

# Print the results
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Salary Range:", salary_range)
```

```python
#Question 3

# Add an outlier (extremely high salary)
outlier_name = "Joseph"
outlier_salary = 1000000
df_with_outlier = df.append({'Gender': "M", 'Name ': outlier_name, 'Annual Salary($)': outlier_salary}, ignore_index=True)

# Calculate the mean, median, and mode with the outlier
mean_salary_with_outlier = df_with_outlier['Annual Salary($)'].mean()
median_salary_with_outlier = df_with_outlier['Annual Salary($)'].median()

# Print the results
print("Mean Salary (without outlier):", mean_salary)
print("Median Salary (without outlier):", median_salary)
print("Mean Salary (with outlier):", mean_salary_with_outlier)
print("Median Salary (with outlier):", median_salary_with_outlier)
```

```python
#Question 4 :
# Calculate the variance of annual salaries for all employees
variance_all = df['Annual Salary($)'].var()

# Calculate the standard deviation of annual salaries for all employees
std_dev_all = df['Annual Salary($)'].std()

# Calculate the variance and standard deviation of annual salaries for male employees
variance_male = df[df['Gender'] == 'M']['Annual Salary($)'].var()
std_dev_male = df[df['Gender'] == 'M']['Annual Salary($)'].std()

# Calculate the variance and standard deviation of annual salaries for female employees
variance_female = df[df['Gender'] == 'F']['Annual Salary($)'].var()
std_dev_female = df[df['Gender'] == 'F']['Annual Salary($)'].std()

# Print the results
print("Variance  (All employees):", variance_all)
print("Standard deviation (All employees):", std_dev_all)
print()
print("Variance (Male employees):", variance_male)
print("Standard deviation (Male employees):", std_dev_male)
print()
print("Variance (Female employees):", variance_female)
print("Standard deviation (Female employees):", std_dev_female)
```

```python
#Question 5 :
# Calculate the IQR for annual salary
q1 = df['Annual Salary($)'].quantile(0.25)
q3 = df['Annual Salary($)'].quantile(0.75)
iqr = q3 - q1

# Print the result
print("Interquartile Range (IQR) for Annual Salary:", iqr)

```
