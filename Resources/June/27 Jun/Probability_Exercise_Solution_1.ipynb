```python
import math
```

```python
# 1. A marketing agency is conducting a survey among a group of 100 consumers. There are 60 males and 40 females in the group.
# The agency wants to randomly select 5 consumers to participate in an in-depth interview. What is the probability that at least 2 of the selected consumers are females?

total_consumers = 100
females = 40

# Function to calculate the combination (nCr)
def nCr(n, r):
    return math.factorial(n) / (math.factorial(r) * math.factorial(n - r))

# Calculate the probability of selecting at least 2 females
probability = 0
for i in range(2, 6):
    probability += (nCr(females, i) * nCr(total_consumers - females, 5 - i)) / nCr(total_consumers, 5)
print("Probability of selecting at least 2 female consumers:", round(probability,3))
```

```python
# 2. An anti-aircraft gun can take a maximum of 4 shots at an enemy plane moving away from it. The probabilities of hitting the plane
#  at the 1st, 2nd, 3rd, and 4th shots are 0.4, 0.3, 0.2 and 0.1. What is the probability that the gun strikes the plane?

# Probabilities of hitting the plane at each shot
probabilities = [0.4, 0.3, 0.2, 0.1]

# Initialize the total probability
total_probability = 0

# Iterate over all possible combinations of hits and misses
for i in range(1, 5):
    # Generate all combinations of hits and misses using itertools
    from itertools import combinations

    # Generate all combinations of i hits among the four shots
    hit_combinations = combinations(range(4), i)

    # Calculate the probability for each combination
    for hits in hit_combinations:
        # Calculate the probability of this combination
        probability = 1  # Probability of all hits
        for j in range(4):
            if j in hits:
                probability *= probabilities[j]  # Probability of hit
            else:
                probability *= (1 - probabilities[j])  # Probability of miss

        # Add the probability of this combination to the total
        total_probability += probability

print("Probability that the gun strikes the plane:", round(total_probability,3))
```

```python
# 3. An electronic assembly consists of two sub-systems say S1 and S2. According to prior testing processes, the likelihood of S1 failing is
# 0.2, the likelihood of S2 failing by itself is 0.15 and the likelihood of both S1 and S2 failing is 0.11.
# What is the likelihood that sub-system S1 breaks on its own?

p_s1 = 0.2
p_s1_s2 = 0.11
p_s1_alone = p_s1 - p_s1_s2
print("The likelihood of sub-system S1 breaking on its own: ", round(p_s1_alone,3))
```

```python
# 4. A lab test is 99% effective in detecting IBD when it is in fact present. The test gives a FP (False positive) result for 0.5%
#  of healthy individuals tested. If IBD affects 0.1% of the population, what is the probability that a person
#  has the disease given that his/her test result is positive?

# Define probabilities
P_IBD = 0.001  # Probability of having IBD in the population
P_not_IBD = 0.999  # Probability of not having IBD in the population
P_Positive_given_IBD = 0.99  # Probability of a positive test result given IBD
P_Positive_given_not_IBD = 0.005  # Probability of a positive test result given not IBD

# Calculate P(Positive)
P_Positive = (P_Positive_given_IBD * P_IBD) + (P_Positive_given_not_IBD * P_not_IBD)

# Calculate P(IBD|Positive) using Bayes' theorem
P_IBD_given_Positive = (P_Positive_given_IBD * P_IBD) / P_Positive

print("Probability that a person has the disease given a positive test result:", round(P_IBD_given_Positive,3))
```

```python
# 5. An insurance company insured 3000 bikes, 4000 cars and 5000 trucks. The probability of an accident involving a bike, a car, and
# a truck are 0.02, 0.03, 0.04 respectively. Find the probability that an insured bike is involved in an accident.

P_Bike = 3000 / (3000 + 4000 + 5000)  # Probability of selecting a bike
P_Accident_given_Bike = 0.02          # Probability of an accident involving a bike

# Calculate the probability that an insured bike is involved in an accident
P_Bike_and_Accident = P_Bike * P_Accident_given_Bike

print("Probability that an insured bike is involved in an accident:", P_Bike_and_Accident)
```
