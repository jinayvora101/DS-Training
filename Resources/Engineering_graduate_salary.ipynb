```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics

pd.set_option('display.max_columns', 500)
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv("July/3 Jul/Engineering_graduate_salary.csv")
df
```

```python
df.info()
```

```python
df2 = df[["Gender", "10percentage", "CollegeTier", "collegeGPA", "ComputerScience", "conscientiousness", "agreeableness", "nueroticism", "English", "Logical", "Quant", "Salary"]]
df2
```

```python
df2["Gender"] = df2["Gender"].map({"m": 1, "f": 0})
df2["ComputerScience"] = df2["ComputerScience"].apply(lambda x: int(x>0))
df2
```

```python
trainX, testX, trainY, testY = train_test_split(df2.iloc[:, :-1], df2.iloc[:, -1], train_size=0.75, test_size=0.25, random_state=69)
regr = LinearRegression()

regr.fit(X=trainX, y=trainY)

regr.coef_, regr.intercept_
```

```python
import statsmodels.api as sm

# trainX = sm.add_constant(trainX)
# testX = sm.add_constant(testX)

model1 = sm.OLS(trainY,trainX)
results1 = model1.fit()
print(results1.summary())
```

```python
predictedY = regr.predict(testX)
```

```python
sklearn.metrics.r2_score(testY, predictedY)
```

```python
error = testY - regr.predict(testX)
error = (error - error.mean())/error.std()
sns.histplot(error, binwidth=0.25)
```

```python
error.skew(), error.kurtosis()
```
