```python
#Q15
# H0: There is no significant difference in the choice of GPS providers between regular and market bookings
# H1: There is a significant difference in the choice of GPS providers between regular and market bookings

import scipy.stats as stats
contingency_table = pd.crosstab(df['GpsProvider'], df['Market/Regular '])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
alpha = 0.05

print('p-value: ', p_value)
if p_value < alpha:
    print("There is a significant difference in the choice of GPS providers between regular and market bookings.")
else:
    print("There is no significant difference in the choice of GPS providers between regular and market bookings.")
```

```python
#Q16
plt.figure(figsize=(12, 8))
sns.boxplot(x='vehicleType', y='TRANSPORTATION_DISTANCE_IN_KM', data=df)
plt.xlabel('Vehicle Type')
plt.ylabel('Transportation Distance (in km)')
plt.title('Total Transportation Distance across Different Vehicle Types')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q17
top_gps_providers = df['GpsProvider'].value_counts().head(10)
total_gps_providers = len(df['GpsProvider'])
shares = (top_gps_providers / total_gps_providers) * 100
gps_provider_data = pd.DataFrame({'GPS Provider': top_gps_providers.index, 'Count': top_gps_providers.values, 'Share (%)': shares.values})
gps_provider_data = gps_provider_data.sort_values(by='Count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(gps_provider_data['GPS Provider'], gps_provider_data['Share (%)'], color='b')
plt.xticks(rotation=45, ha='right')
plt.xlabel('GPS Provider')
plt.ylabel('Shares of Providers')
plt.title('Top 10 GPS Providers and Their Respective Shares')
plt.show()
```

```python
#Q18
booking_counts = df['Market/Regular '].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(booking_counts, labels=booking_counts.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Distribution of Market and Regular Bookings')
plt.show()
```

```python
#Q19
regular_bookings = df[df['Market/Regular '] == 'Regular']
market_bookings = df[df['Market/Regular '] == 'Market']
regular_trips_by_month = regular_bookings['Month'].value_counts().sort_index()
market_trips_by_month = market_bookings['Month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(regular_trips_by_month.index, regular_trips_by_month.values, label='Regular Bookings', marker='o')
plt.plot(market_trips_by_month.index, market_trips_by_month.values, label='Market Bookings', marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.title('Number of Monthly Trips for Regular and Market Bookings')
plt.legend()
plt.xticks(ticks=range(1, 13), labels=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.grid(True)
plt.show()
```
