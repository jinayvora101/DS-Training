# How to Handle Categorical Data?

Real world data comes with their unique blends. Sometime working with real world data, you will have to deal with categorical data and other time not. Categorical data are those types of data whose features' values contain limited number of categories.

In many cases, categorical features have text values. And most ML models accept numerical inputs. That is the reason why we have to manipulate these types of categories to be in proper format accepted by ML algorithms.

There are four techniques to encode or convert the categorical features into numbers. Here are them:

* [Mapping Method](#1)
* [Ordinal Encoding](#2)
* [Label Encoding](#3)
* [Pandas Dummies](#4)
* [OneHot Encoding](#5)

Note that some of these encoding techniques can produce same output, the difference is only implementation. The first 3 will produce the numerical outputs while the latter will produce the one hot matrix (with 1s and 0s).

Let's implement them.

## 1. Mapping Method

Mapping method is straight forward way to encode categorical features with few categories.

Let's apply it to the class feature: It has three categories: `Third, First, Second`. We create a dictionary whose keys are categories and values are numerics to encode to and then map it to the dataframe.

Here is how it is done:

```python
map_dict = {
    'First':0,
    'Second': 1,
    'Third': 2
}
```

```python
titanic_df = titanic.copy()
```

```python
titanic_df['class'] = titanic_df['class'].map(map_dict)
```

```python
titanic_df.head()
```

```python
titanic_df['class'].value_counts()
```

<a name='2'></a>
## 2. Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

cats_feats = titanic_df[['alive', 'alone']]

encoder = OrdinalEncoder()

cats_encoded = encoder.fit_transform(cats_feats)
```

The output of the encoder is a NumPy array. We can convert it back to the pandas dataframe.

```python
titanic_df[['alive', 'alone']] = pd.DataFrame(cats_encoded, columns=cats_feats.columns, index=cats_feats.index)
titanic_df.head()
```

```python
encoder.categories_
```

**Warning**: Ordinal Encoder can't handle missing values. It will throw an error. You may try it on `embarked` and see.

<a name='3'></a>
## 3. Label Encoding

Label Encoding is noted to used for encoding target features [(as per sklearn documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html?highlight=label%20encoder#sklearn.preprocessing.LabelEncoder) but otherwise, it can also be used to achieve our purpose of encoding categorical features.

It also can't support missing values. So to make it simple, let's drop all missing values.

```python
titanic_cleaned = titanic.dropna()
```

```python
titanic_cleaned.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

deck_feat = titanic_cleaned[['deck']]

label_encoder = LabelEncoder()

deck_encoded = label_encoder.fit_transform(deck_feat)
```

Similar to ordinal encoder, the output of Label Encoder is a NumPy array.

```python
titanic_cleaned['deck'] = pd.DataFrame(deck_encoded, columns=deck_feat.columns, index=deck_feat.index)

titanic_cleaned.head()
```

```python
label_encoder.classes_
```

```python
titanic_cleaned['deck'].value_counts()
```

<a name='4'></a>
## 4. Pandas Dummies

This is another easy way to handle categorical features. It will create extra features based on the available categories. Let's apply it to the feature `who`.

```python
dummies = pd.get_dummies(titanic['who'], drop_first=True)
```

```python
titanic_dummy = pd.concat([titanic.drop('who',axis=1),dummies],axis=1)
```

```python
titanic_dummy.head()
```

<a name='5'></a>
## 5. One Hot Encoding

This method will convert a feature into one hot matrix. Additional features corresponding to the values of the given categories will be created.

Basically same as dummies.

```python
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder()

town_encoded = one_hot.fit_transform(titanic_cleaned[['embark_town']])
```

```python
one_hot.categories_
```

```python
town_encoded
```

The output of One hot encoder is a sparse matrix. We will need to convert it into NumPy array.

```python
town_encoded = town_encoded.toarray()
```

```python
columns = list(one_hot.categories_)

town_df = pd.DataFrame(town_encoded, columns =columns)

town_df.head()
```

```python
len(town_df)
```

```python
len(titanic_cleaned)
```

```python
drop_embark = titanic_cleaned.drop('embark_town',axis=1)

drop_embark[['Cherbourg', 'Queenstown', 'Southampton']] = town_df
```
