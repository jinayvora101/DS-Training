```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df["Survived"].value_counts().plot(kind="pie")
```

```python
df.groupby("Pclass").agg({"Fare": np.mean})
```

```python
df.query(""" Embarked == "C" """).count()
```

```python
df["Fare"].plot.box()
```

```python
df.query(""" Survived == 1 """)["Age"].plot.hist()
```

```python
df.query(""" Survived == 0 """)["Age"].plot.hist()
```

```python
df[["Sex", "Survived"]].value_counts().plot(kind="bar", stacked=True)
```

```python
df.query(""" SibSp >= 4 """).sort_values(by="Fare", ascending=False)
```

```python
df.plot(x="Age", y="Fare", kind="scatter")
```

```python
df.corr(method="pearson").round(4)*100
```

```python
pd.pivot_table(data=df, index="Sex", columns="Embarked", values="Survived", aggfunc="count")
```
