1. How many people survived the titanic crash? Can you use appropriate visualization(s) to showcase this?

#### Bar Plot

A bar plot presents bars side by side, representing different sub-categories or variables.

```python
survived_counts = titanic_df['survived'].value_counts()
```

```python
sns.set_palette("pastel")      # setting the overall color palette
bar = sns.barplot(x=survived_counts.index, y=survived_counts.values)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Rate of Titanic Crash')
plt.xticks([0, 1], ['Did not survive', 'Survive'])
plt.show()
```

Survival rate is on the lower spectrum

4. Visualize the gender ratio of passengers aboard the Titanic.

#### Pie Chart
A pie chart is a pictorial representation of data in the form of a circular chart or pie where the slices of the pie show the size of the data.

```python
passenger_counts = titanic_df['sex'].value_counts()
sns.set(style="darkgrid")
plt.figure(figsize=(6, 6))
plt.pie(passenger_counts, labels=passenger_counts.index, autopct='%1.1f%%', startangle=90)
```

The male passengers are more (almost double) than the females.

#### Donut Chart
- A donut chart is a pie chart with part of the centre removed.
- Pie charts are sometimes criticized for drawing readers' attention to how the slices' areas relate to one another and to the entire chart. This makes it challenging to distinguish between slices, especially when comparing multiple Pie Charts at once.
- Downplaying the use of the space, a donut chart somewhat addresses this issue. Instead of comparing the ratios between slices, readers place more emphasis on reading the length of the arcs.
- Donut Charts are also more space-efficient than Pie Charts because they allow for the display of data inside their empty spaces.




```python
plt.figure(figsize=(10,8))
plt.pie(passenger_counts, labels=passenger_counts.index, wedgeprops={'width':0.3}, startangle=90, autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

#### Histogram
- A histogram is a visual representation of the distribution of a continuous variable.
- It provides insights into the shape, central tendency, and spread of the data.


8. What is the underlying distribution of passenger age?

```python
sns.histplot(data=titanic_df, x='age', color="purple")
```

The passengers are mostly from 20-30 age bracket and the underlying distribution is somewhat positively skewed.


#### Density Plots
A density plot may be seen as an extension of the histogram. As opposed to the histogram, the density plot can smooth out the distribution of values and reduce the noise.

```python
# Set up the figure and axes
fig, ax = plt.subplots()

# Density plot using Seaborn
sns.kdeplot(data=titanic_df, x="age", fill=True, ax=ax)
ax.set_title('Age group distribution')
ax.set_xlabel('Age')
ax.set_ylabel('Density')
```

#### Add Kernal Density Estimate
- Kernel Density Estimation (KDE) is a non-parametric technique used to estimate the probability density function (PDF) of a random variable.
- Setting kde=True, the histplot() function will add a smoothed KDE line on top of the histogram, providing an estimate of the underlying probability density function. This allows for a more comprehensive visualization of the data distribution.


```python
# Histogram with kernel density estimate
sns.histplot(data=titanic_df, x='age', kde=True)
plt.title('Age group distribution')
plt.show()
```

#### Adding bins
- The bins parameter on a histplot divides the range of values of a continuous variable into discrete intervals or groups.
- These intervals or bins help to organize and summarize the data in a histogram by grouping similar values together.


```python
sns.histplot(data=titanic_df, x='age', color="purple", bins=5, kde=True)
```

9. How does the passenger age and class interact with each other?

#### Box and Whiskers Plot
- A box and whiskers plot, also called a box plot, helps in visualizing the distribution of numerical data.
- The whiskers extend from the box to show the range of the data, excluding outliers.
- Box plots provide information about the central tendency, spread, skewness, and presence of outliers in the data distribution.

12. Analyze the correlation between all the attributes of climate data.

#### Heatmap
- A heat map is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions.
- The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varied over space

```python
# Calculate the correlation matrix
numeric_feats = climate_df.select_dtypes(include='number')
corr_matrix = numeric_feats.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True)
plt.show()
```
